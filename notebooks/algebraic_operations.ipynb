```python
from tensors import Tensor, Covector, Vector, tensor_product, tensor_sum, tensor_contract
import numpy as np
```

#### Scalar Multiplication

```python
# On Vector
scalar: float = 3.0
v = Vector.from_data([1.0, 2.0, -1.0])
pre_scaled_v = scalar*v
post_scaled_v = v*scalar
print(v)
print(pre_scaled_v)
print(post_scaled_v)


# On Covector
scalar2: float = 3.0
alpha = Covector.from_data([2.0, 3.0, -1.0])
pre_scaled_alpha = scalar2*alpha
post_scaled_alpha = alpha*scalar2
print(alpha)
print(pre_scaled_alpha)
print(post_scaled_alpha)
```

```python
# Addition
M = np.array([[1, 3], [5, 1]])
T = Tensor.from_data(M, tensor_type = (1, 1))
print(T)
print(v+pre_scaled_v)
print(alpha+pre_scaled_alpha)

```

```python
alpha = Covector.from_data([4, 5, 6])
v = Vector.from_data([1, 2, 3])

print(alpha.evaluate(v))  
print(alpha(v))           
```

### Contraction

Action of a Tensor(1,1) on a Vector via Contraction

```python
T = Tensor.from_data([[1, 3], [5, 1]], tensor_type = (1, 1))
V = Vector.from_data([2, 4])
print(T)
print(V)
W = tensor_contract(
    T,
    V,
    contravariant_axis=("right", 0),
    covariant_axis=("left", 1),
)
print(W)
```

Action of a Tensor(1,1) on a Covector via Contraction

```python
T = Tensor.from_data([[1, 3], [5, 1]], tensor_type = (1, 1))
alpha = Covector.from_data([2, 4])
print(T)
print(alpha)
B = tensor_contract(
    T,
    alpha,
    contravariant_axis=("left", 0),
    covariant_axis=("right", 0),
)
print(B)
```

Tensor (2, 2). Components:

$A^{ij}{}_{kl} \leftrightarrow \texttt{data[i, j, k, l]} $

Case:

$
A^{ij}{}_{kl} =
\begin{cases}
1 & \text{si } i=j=k=l=1 \\
2 & \text{si } i=1,\ j=2,\ k=1,\ l=2 \\
3 & \text{si } i=2,\ j=1,\ k=2,\ l=1 \\
4 & \text{si } i=j=k=l=2 \\
0 & \text{en otro caso}
\end{cases}
$

```python
data = np.zeros((2, 2, 2, 2))

# A^{11}_{11} = 1
data[0, 0, 0, 0] = 1

# A^{12}_{12} = 2
data[0, 1, 0, 1] = 2

# A^{21}_{21} = 3
data[1, 0, 1, 0] = 3

# A^{22}_{22} = 4
data[1, 1, 1, 1] = 4

A = Tensor.from_data(data, tensor_type=(2, 2))
print(A)
```

Contract $j$ with $l$:

$
B^{i}{}_{k} = A^{ij}{}_{kj} = \sum_{j=1}^{2} A^{ij}{}_{kj}
$

```python
# B^i_k = A^{ij}_{kj}
contracted_data = np.einsum("ijkj->ik", A.components)

B = Tensor.from_data(contracted_data, tensor_type=(1, 1))

print(B)
```

$$
B^{i}{}_{k}
=
\begin{pmatrix}
3 & 0 \\
0 & 7
\end{pmatrix}
$$
