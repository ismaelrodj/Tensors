```python
from tensors import Tensor, Covector, Vector
import numpy as np
```

```python
# Scalar multiplication
scalar: float = 3.0
v = Vector.from_data([1.0, 2.0, -1.0])
pre_scaled_v = scalar*v
post_scaled_v = v*scalar
print(v)
print(pre_scaled_v)
print(post_scaled_v)


# Scalar multiplicationk on Covector
scalar2: float = 3.0
alpha = Covector.from_data([2.0, 3.0, -1.0])
pre_scaled_alpha = scalar2*alpha
post_scaled_alpha = alpha*scalar2
print(alpha)
print(pre_scaled_alpha)
print(post_scaled_alpha)
```

```python
# Addition
M = np.array([[1, 3], [5, 1]])
T = Tensor.from_data(M, tensor_type = (1, 1))
print(T+T)
print(v+pre_scaled_v)
print(alpha+pre_scaled_alpha)

```

## Tensor product as a matrix view

For higher-rank tensors, we can regroup indices into rows and columns to inspect the components in a more ordered way.


```python
# Tensor product
tensor_product_result = T @ v
print(tensor_product_result)

tensor_product_result.display_matrix(row_axes=(0, 1), col_axes=(2,))
tensor_product_result.display_slice(0)
```
